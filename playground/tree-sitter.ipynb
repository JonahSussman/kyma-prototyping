```python
from tree_sitter import Language, Parser

Language.build_library(
  "build/language-java.so",
  ["../tree-sitter-java/"]
)

LANGUAGE_JAVA = Language("build/language-java.so", "java")

parser = Parser()
parser.set_language(LANGUAGE_JAVA)
```

```python
FILEPATH = "ComplexA.java"

with open(FILEPATH) as f:
  src = f.read()
```

```python
tree = parser.parse(bytes(src, "utf8"))
```

```python
def format_s_expression(s):
  indent_size = 2
  level = 0
  formatted = ""
  i = 0

  while i < len(s):
    char = s[i]

    if char == '(':
      if i > 0 and s[i - 1] != '(':
        formatted += '\n' + ' ' * (level * indent_size)
      formatted += char
      level += 1
    elif char == ')':
      level -= 1
      if i > 0 and s[i - 1] != '(' and s[i - 1] != ')':
        formatted += '\n' + ' ' * (level * indent_size)
      formatted += char
    else:
      if i > 0 and s[i - 1] == '(':
        formatted += '\n' + ' ' * (level * indent_size)
      formatted += char

    i += 1

  return formatted
```

```python
print(format_s_expression(tree.root_node.sexp()))
```
